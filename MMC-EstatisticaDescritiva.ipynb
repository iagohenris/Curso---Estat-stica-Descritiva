```python
import random
```

```python
T = 9000000 #Número de Tentativas do sorteio
```

```python
#Quanto maior o valor de T maior a precisão do valor de Pi
```

```python
I = 0
for i in range(T):
  x = random.random()
  y = random.random()
  if(((x*x) +(y*y)) < 1):
    I = I + 1
```

```python
I
```

```python
Pi = 4*I/T
Pi
```
