<h1> Gráficos de colunas / barras <h1>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
y = [2, 5, 2, 7, 5, 1]
x = ['N1', 'N2', 'N3', 'N4', 'N5', 'N6']
x2 = ['Variável 1', 'Variável 2', 'Variável 3', 'Variável 4', 'Variável 5', 'Variável 6']
```

```python
# Barras horizontais
```

```python
plt.barh(x2,y, color="r")
plt.xlabel('Variável eixo X', size=24)
plt.ylabel('Categorias', size=20)
plt.title('Título do meu gráfico')
```

```python
# Barras verticais
```

```python
plt.bar(x,y, color='purple')
```

```python
sns.barplot(x,y)
```

```python
sns.barplot(x2,y)
```

```python
plt.pie(y, labels=x, radius=1)
plt.show()
```

```python
y = [6, 8, 3, 1, 9]
x1 = [1, 2, 3, 4, 5]
x = ['seg', 'ter', 'qua', 'qui', 'sex']
```

```python
plt.plot(x1, y, "o-g")
plt.xlabel('Eixo X', size=20)
plt.ylabel('Eixo Y', size=20)
plt.title('Título do gráfico', size=15)
plt.show()
```

```python
plt.plot(x, y, "o-c")
plt.xlabel('Eixo X', size=20)
plt.ylabel('Eixo Y', size=20)
plt.title('Título do gráfico', size=15)
plt.show()
```
